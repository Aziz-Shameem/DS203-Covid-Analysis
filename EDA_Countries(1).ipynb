```python
df_corr_india=h[h['country']=="India"]['active_cases']
df_corr_india.reset_index(inplace=True, drop=True)
df_corr_usa=h[h['country']=="USA"]['active_cases']
df_corr_usa.reset_index(inplace=True, drop=True)
df_corr_italy=h[h['country']=="Italy"]['active_cases']
df_corr_italy.reset_index(inplace=True, drop=True)
df_corr_germany=h[h['country']=="Germany"]['active_cases']
df_corr_germany.reset_index(inplace=True, drop=True)
df_corr_ukraine=h[h['country']=="Ukraine"]['active_cases']
df_corr_ukraine.reset_index(inplace=True, drop=True)

df_corr_mexico=h[h['country']=="Mexico"]['active_cases']
df_corr_mexico.reset_index(inplace=True, drop=True)
df_corr_indonesia=h[h['country']=="Indonesia"]['active_cases']
df_corr_indonesia.reset_index(inplace=True, drop=True)
df_corr_turkey=h[h['country']=="Turkey"]['active_cases']
df_corr_turkey.reset_index(inplace=True, drop=True)
df_corr_brazil=h[h['country']=="Brazil"]['active_cases']
df_corr_brazil.reset_index(inplace=True, drop=True)
```

```python
data = [df_corr_india, df_corr_usa, df_corr_italy, 
        df_corr_germany, df_corr_ukraine,df_corr_mexico, df_corr_indonesia, df_corr_turkey, df_corr_brazil]


headers = ['India','USA','Italy','Germany','Ukraine','Mexico','Indonesia','Turkey','Brazil']

df3_correlation= pd.concat(data, axis=1,keys=headers)

fig=plt.figure(figsize=(10,8))
sns.heatmap(df3_correlation.corr(), cmap='YlGnBu',annot=True)
plt.title('Correlation matrix for daily (Active) cases', weight='bold', fontsize=20)
#plt.show()

#plt.savefig(r'C:\Users\Aziz_Shameem\Documents\heatmaps\active.png')
```
